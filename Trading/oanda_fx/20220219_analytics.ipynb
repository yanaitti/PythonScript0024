```python
df_ohlcv = pd.read_csv('datas/USDJPY_M1_201601040000_202202182359.csv', sep='\t', names=['date', 'time', 'open', 'high', 'low', 'close', 'tickvol', 'vol','spread'], skiprows=1)
df_ohlcv
```

```python
df_ohlcv['datetime'] = df_ohlcv['date'] + ' ' + df_ohlcv['time']
df_ohlcv['datetime'] = pd.to_datetime(df_ohlcv['datetime'], format='%Y.%m.%d %H:%M:%S')
df_ohlcv['weekday'] = df_ohlcv['datetime'].dt.weekday

df_ohlcv['weekday'] = df_ohlcv['datetime'].dt.weekday

df_ohlcv['sma20'] = ta.SMA(df_ohlcv['close'], timeperiod=20)
df_ohlcv['sma50'] = ta.SMA(df_ohlcv['close'], timeperiod=50)
df_ohlcv['sma100'] = ta.SMA(df_ohlcv['close'], timeperiod=100)

df_ohlcv['macd'], df_ohlcv['macdsignal'], df_ohlcv['macdhist'] = ta.MACD(df_ohlcv['close'], fastperiod=12, slowperiod=26, signalperiod=9)
df_ohlcv['bb_2_upper'], df_ohlcv['bb_2_mid'], df_ohlcv['bb_2_lower'] = ta.BBANDS(df_ohlcv['close'], timeperiod=21, nbdevup=2, nbdevdn=2, matype=0)
df_ohlcv['macd'], df_ohlcv['macdsignal'], df_ohlcv['macdhist'] = ta.MACD(df_ohlcv['close'], fastperiod=12, slowperiod=26, signalperiod=9)
df_ohlcv['slowk'], df_ohlcv['slowd'] = ta.STOCH(df_ohlcv['high'], df_ohlcv['low'], df_ohlcv['close'], fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)

df_ohlcv['diff_hc'] = df_ohlcv['high'] - df_ohlcv['close']
df_ohlcv['diff_cl'] = df_ohlcv['close'] - df_ohlcv['low']

df_ohlcv
```
