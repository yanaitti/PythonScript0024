```python
data = pd.read_csv('inputs/USDJPY_M15.csv', sep='\t', names=('date', 'time', 'open', 'high', 'low', 'close'), usecols=[0, 1, 2, 3, 4, 5], skiprows=1)
# data = pd.read_csv('inputs/USDJPY15.csv', names=('date', 'time', 'open', 'high', 'low', 'close'), usecols=[0, 1, 2, 3, 4, 5], skiprows=1)
data['datetime'] = pd.to_datetime(data['date']  + ' ' + data['time'])
data.drop(['date', 'time'], axis=1, inplace=True)
data['time'] = data['datetime'].apply(lambda x: int(x.timestamp()))
data.drop(['datetime'], axis=1, inplace=True)
data['currency'] = 'USDJPY'
data['period'] = '15'
data['id'] = ''
data = data[:-1]
data
```
